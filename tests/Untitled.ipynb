```python
def calculate_coindcx_order(order_value, trading_fee_percent=0.5, gst_percent=18, tds_percent=0):
    """
    Calculate the total amount payable for a CoinDCX order.

    Parameters:
    order_value (float): The value of the order in INR.
    trading_fee_percent (float): The trading fee percentage (default is 0.5%).
    gst_percent (float): The GST percentage on the trading fee (default is 18%).
    tds_percent (float): The TDS percentage (default is 0%).

    Returns:
    dict: A dictionary with detailed cost breakdown.
    """
    trading_fee = (trading_fee_percent / 100) * order_value
    gst = (gst_percent / 100) * trading_fee
    tds = (tds_percent / 100) * order_value  # Currently 0% for INR buys

    total_payable = order_value + trading_fee + gst

    return {
        "Order Value": order_value,
        "Trading Fee (0.5%)": trading_fee,
        "GST on Fee (18%)": gst,
        "TDS (0%)": tds,
        "Total Payable": total_payable
    }

# Example usage
order_summary = calculate_coindcx_order(5000.00)
for key, value in order_summary.items():
    print(f"{key}: {value:.2f}")
```

```python
import time
import requests
import hmac
import hashlib
import pandas as pd
from datetime import datetime
import logging
import json

from services.coindcx_api import CoinDCXAPI
k=CoinDCXAPI()

# === CONFIGURATION ===
# API_KEY = "f0953d53d66d2909d0e24e45a692b5066c907d2530d16a7f"  # Replace with your API key
API_KEY=k.api_key
# API_SECRET = "2292ac809896e56bb159a2516d37ec679ac2dccebbc86b0de49fbd2cd749ce95"
API_SECRET=k.api_secret
BASE_URL = "https://api.coindcx.com"
TRADE_PAIR = "VINUINR"
FIXED_INVESTMENT = 110  # INR investment per trade
CSV_FILE = "market_data.csv"
RSI_PERIOD = 14
SMA_PERIOD = 50
PROFIT_TARGET = 1.05  # 5% target profit
STOP_LOSS = 0.98  # 2% stop loss
RISK_PER_TRADE = 0.02  # 2% of total capital per trade
ORDER_STATUS_CHECK_DELAY = 5  # Seconds to wait before checking order status

# === LOGGING ===
logging.basicConfig(filename="trading_bot.log", level=logging.INFO,
                    format="%(asctime)s - %(levelname)s - %(message)s")

# === API AUTHENTICATION ===
def generate_signature(payload, secret):
    return hmac.new(secret.encode(), payload.encode(), hashlib.sha256).hexdigest()

# def get_account_balance(currency="INR"):
def get_account_balance(symbol="INR"):
    return float(k.get_wallet_balance())

    # from services.coindcx_api import CoinDCXAPI
# k=CoinDCXAPI()
    # BASE_URL = "https://api.coindcx.com"
    # endpoint = "/exchange/v1/users/balances"
    # url = BASE_URL + endpoint
    # timestamp = int(time.time() * 1000)
    # payload = {
    #     "timestamp": timestamp
    # }
    # json_payload = json.dumps(payload, separators=(',', ':'))
    # signature = generate_signature(json_payload,API_SECRET)
    # headers = {
    #     "Content-Type": "application/json",
    #     "X-AUTH-APIKEY": API_SECRET,
    #     "X-AUTH-SIGNATURE": signature
    # }
    # try:
    #     response = requests.post(url, data=json_payload, headers=headers)
    #     response.raise_for_status()  # Check for HTTP errors (4xx or 5xx)
    #     balances = response.json()
    #     # return pd.DataFrame(balances)          
    #     for balance in balances:
    #         if balance['currency'] == symbol[:3]:       
    #             # if balance['locked_balance']:
    #             #     return f"Locked {balance['currency']} at {balance['locked_price']}"
    #             # else:
    #             return float(balance['balance'])
    #     return None
    # except requests.exceptions.RequestException as e:
    #     print(f"Error during API request: {e}")
    #     return None

# === FETCH MARKET DATA ===
def get_latest_price(pair):
    try:
        response = requests.get(f"{BASE_URL}/exchange/ticker")
        response.raise_for_status()
        prices = response.json()
        for market in prices:
            if market["market"] == pair:
                return float(market["ask"]), float(market["bid"])
        logging.error(f"Pair {pair} not found in market data.")
        return None, 
        None
    except requests.exceptions.RequestException as e:
        logging.error(f"Error fetching market data: {e}")
        return None, None

# === SAVE DATA TO CSV ===
def save_to_csv(data):
    df = pd.DataFrame(data, columns=["timestamp", "ask_price", "bid_price"])
    df.to_csv(CSV_FILE, mode='a', index=False, header=not pd.io.common.file_exists(CSV_FILE))

# === CALCULATE INDICATORS ===
def calculate_indicators(df):
    df["ask_price"] = pd.to_numeric(df["ask_price"], errors='coerce')
    df["sma"] = df["ask_price"].rolling(window=SMA_PERIOD).mean()
    delta = df["ask_price"].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=RSI_PERIOD).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=RSI_PERIOD).mean()
    rs = gain / loss if loss.all() != 0 else 0
    df["rsi"] = 100 - (100 / (1 + rs))
    return df

# === CHECK TREND AND RSI ===
def check_trend_and_rsi(df):
    if len(df) < SMA_PERIOD:
        return False, False, False
    is_uptrend = df["ask_price"].iloc[-1] > df["sma"].iloc[-1]
    rsi = df["rsi"].iloc[-1]
    is_overbought = rsi > 70
    is_oversold = rsi < 30
    return is_uptrend, is_overbought, is_oversold

# === CHECK FUNDS ===
def check_funds(quantity, price):
    try:
        order_value = quantity * price
        order_summary = calculate_coindcx_order(order_value)
        total_payable = order_summary["Total Payable"]  # Get total payable with fees
        available_balance = get_account_balance("INR")
        return available_balance >= total_payable
    except Exception as e:
        logging.error(f"Error checking funds: {e}")
        return False

# === PLACE MARKET ORDER ===
def place_order(side, quantity, price):
    endpoint = "/exchange/v1/orders/create"
    payload = {
        "side": side,
        "order_type": "market_order",
        "market": TRADE_PAIR,
        "price_per_unit": str(price),
        "total_quantity": str(quantity)
    }

    payload_str = str(payload).replace("'", '"')
    signature = generate_signature(payload_str, API_SECRET)

    headers = {
        "X-AUTH-APIKEY": API_KEY,
        "X-AUTH-SIGNATURE": signature,
        "Content-Type": "application/json"
    }

    try:
        response = requests.post(BASE_URL + endpoint, json=payload, headers=headers)
        response.raise_for_status()
        order_details = response.json()
        logging.info(f"{side.upper()} order placed: {order_details}")
        return order_details
    except requests.exceptions.RequestException as e:
        logging.error(f"Error placing order: {e}")
        return {"status": "error", "message": str(e)}

def calculate_coindcx_order(order_value, trading_fee_percent=0.5, gst_percent=18, tds_percent=0):
    """
    Calculate the total amount payable for a CoinDCX order.

    Parameters:
    order_value (float): The value of the order in INR.
    trading_fee_percent (float): The trading fee percentage (default is 0.5%).
    gst_percent (float): The GST percentage on the trading fee (default is 18%).
    tds_percent (float): The TDS percentage (default is 0%).

    Returns:
    dict: A dictionary with detailed cost breakdown.
    """
    trading_fee = (trading_fee_percent / 100) * order_value
    gst = (gst_percent / 100) * trading_fee
    tds = (tds_percent / 100) * order_value  # Currently 0% for INR buys

    total_payable = order_value + trading_fee + gst

    return {
        "Order Value": order_value,
        "Trading Fee (0.5%)": trading_fee,
        "GST on Fee (18%)": gst,
        "TDS (0%)": tds,
        "Total Payable": total_payable
    }

# === CHECK ORDER STATUS ===
def check_order_status(order_id):
    # Placeholder - Implement actual API call to check order status
    time.sleep(ORDER_STATUS_CHECK_DELAY)
    status = "filled"  # Placeholder
    logging.info(f"Order status for {order_id}: {status}")
    return status

# === CONSTANTS ===
TAKER_FEE = 0.0005  # CoinDCX taker fee (0.05%)

# === FEE CALCULATION FUNCTIONS ===
def calculate_taker_fee(trade_value):
    """Calculate the taker fee for a trade."""
    return trade_value * TAKER_FEE

def calculate_total_cost(quantity, price):
    """Calculate the total cost of a buy order (including fees)."""
    trade_value = quantity * price
    fee = calculate_taker_fee(trade_value)
    return trade_value + fee

def calculate_net_quantity(quantity):
    """Calculate the net quantity received after fees for a buy order."""
    return quantity * (1 - TAKER_FEE)

def calculate_revenue(quantity, price):
    """Calculate the revenue after fees for a sell order."""
    trade_value = quantity * price
    return trade_value * (1 - TAKER_FEE)

# === UPDATED TRADING BOT LOGIC ===
def trading_bot():
    market_data = []
    capital = FIXED_INVESTMENT / RISK_PER_TRADE
    position_held = False
    order_id = None

    while True:
        print("trading bot working")
        ask_price, bid_price = get_latest_price(TRADE_PAIR)

        if ask_price and bid_price:
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            market_data.append([timestamp, ask_price, bid_price])
            save_to_csv(market_data)

            df = pd.read_csv(CSV_FILE)
            df = calculate_indicators(df)
            is_uptrend, is_overbought, is_oversold = check_trend_and_rsi(df)

            if is_uptrend and not is_overbought and not position_held:
                quantity = capital * RISK_PER_TRADE / ask_price
                if check_funds(quantity, ask_price):  # Check if sufficient funds are available
                    order_value = quantity * ask_price
                    order_summary = calculate_coindcx_order(order_value)
                    total_payable = order_summary["Total Payable"]

                    logging.info(f"Attempting to buy {quantity} {TRADE_PAIR} at {ask_price}, Total Payable: {total_payable:.2f}")
                    buy_response = place_order("buy", quantity, ask_price)

                    if buy_response.get("status") == "success":
                        order_id = buy_response.get("order_id")
                        buy_price = ask_price
                        net_quantity = calculate_net_quantity(quantity)  # Adjust quantity for fees
                        target_price = buy_price * PROFIT_TARGET
                        stop_loss_price = buy_price * STOP_LOSS
                        logging.info(f"Bought at {buy_price}. Net Quantity: {net_quantity}. Target: {target_price}, Stop Loss: {stop_loss_price}")
                        position_held = True
                    else:
                        logging.error(f"Buy order failed: {buy_response.get('message')}")
                else:
                    logging.warning("Insufficient funds to place buy order (including fees).")

            elif position_held:
                _, bid_price = get_latest_price(TRADE_PAIR)
                if bid_price:
                    if bid_price >= target_price or bid_price <= stop_loss_price:
                        side = "sell"
                        logging.info(f"Attempting to sell at {bid_price} ({'Target' if bid_price >= target_price else 'Stop-Loss'})")
                        sell_response = place_order(side, net_quantity, bid_price)  # Use net quantity for sell

                        if sell_response.get("status") == "success":
                            sell_order_id = sell_response.get("order_id")
                            if check_order_status(sell_order_id) == "filled":
                                revenue = calculate_revenue(net_quantity, bid_price)  # Calculate revenue after fees
                                capital = capital + (revenue - (quantity * buy_price))  # Update capital
                                position_held = False
                                order_id = None
                                logging.info(f"Sell order filled. Revenue: {revenue}. Updated capital: {capital}")
                            else:
                                logging.warning(f"Sell order status not filled. Check manually. Status: {check_order_status(sell_order_id)}")

            time.sleep(5)  # Check price every 5 seconds

        time.sleep(10)  # Fetch market data every 10 seconds

# === RUN THE BOT ===
if __name__ == "__main__":
#   print(get_account_balance("INR"))
    try:
        trading_bot()
    except Exception as e:
        logging.exception("An unexpected error occurred:")
        print(f"A critical error occurred: {e}. Check the logs for details.")
```
